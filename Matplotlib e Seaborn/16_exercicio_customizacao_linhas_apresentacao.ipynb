## Exercício - Gráfico de linhas

Considere a seguinte tabela com o número de refeições servidas por ano por uma instituição de caridade.

Ano | Refeições
--- | ---
2010 | 40.139
2011 | 127.020
2012 | 168.193
2013 | 153.115
2014 | 202.102
2015 | 232.897
2016 | 277.912
2017 | 205.350
2018 | 233.389
2019 | 232.797
2020 | 110.023
2021 | 160.350
2022 | 222.789

Construa um gráfico de linhas com as seguintes características:

- Tamanho 14x6
- Linha com espessura 3
- Palheta de cores Set3
- Destaque o impacto da pandemia (2020 e 2021) com um vspan
- Destaque o impacto da pandemia com um texto
- Destaque os pontos de início e fim da série com um scatter plot com as quantidades de refeições servidas
- Remova os eixos y
- Remova os ticks do eixo x
- Remova as bordas do gráfico
- Adicione um título
- A figura deve ser salva numa pasta chamada `imagens` que deve estar no mesmo diretório do Jupyter Notebook



```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

```python
import matplotlib.pyplot as plt 
from cycler import cycler
cores = plt.get_cmap('Set3').colors # cores para o gráfico
ciclo_cores = cycler('color', cores) # ciclo de cores
plt.rc('axes', prop_cycle=ciclo_cores) # aplicar ciclo de cores 
cor_anotacoes = cores[4]
plt.figure(figsize=(14, 6)) # tamanho do gráfico
plt.plot(anos, refeicoes, linewidth=3, zorder=1) # plotar o gráfico

for posicao in(-1, 0):
    plt.scatter(anos[posicao], refeicoes[posicao], color=cor_anotacoes, zorder=2) # destacar os pontos
    plt.annotate(refeicoes[posicao], color=cor_anotacoes, xy=(anos[posicao], refeicoes[posicao]), xytext=(5, 0), textcoords='offset points', size=12, weight='bold') # anotar os valores  

plt.box(False) # remover a caixa    
plt.yticks([]) # remover os valores do eixo y
plt.xticks(anos) 
plt.tick_params(axis='x', length=0) # remover os traços do eixo x
plt.title('Refeições servidas por ano', fontsize=16, weight='bold', loc='left', color='gray') # título do gráfico

plt.axvspan(2020, 2021, alpha=0.5, color=cores[1], zorder=0) # destacar a pandemia
plt.text(2020, 40000, 'Pandemia', fontsize=11, weight='bold', color=cor_anotacoes)
plt.savefig('refeicoes.png', bbox_inches='tight',dpi=170) # salvar o gráfico   
plt.show() # mostrar o gráfico
```
