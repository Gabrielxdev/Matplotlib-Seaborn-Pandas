# Trabalhando com eixos



## Gr√°fico de linhas

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
fig, ax = plt.subplots()  # Cria a figura e os eixos
ax.set_ylabel('M√™s') # Adiciona o r√≥tulo do eixo y
ax.set_xlabel('Vendas') # Adiciona o r√≥tulo do eixo x   
ax.set_title('Vendas ao longo do ano') # Adiciona o t√≠tulo do gr√°fico   

plt.show() # Exibe o gr√°fico    
```

## Gr√°fico de barras

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

fig, ax = plt.subplots()  # Cria a figura e os eixos    

ax.bar(produtos, quantidades) # Cria o gr√°fico de barras    
ax.set_ylabel('Produto') # Adiciona o r√≥tulo do eixo y
ax.set_xlabel('Quantidade de vendas') # Adiciona o r√≥tulo do eixo x   
ax.set_title('quantidade de vendas por produto') # Adiciona o t√≠tulo do gr√°fico 
```

## Gr√°fico de pizza

```python
categorias = ['Eletr√¥nicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

fig, ax = plt.subplots()  # Cria a figura e os eixos    

ax.pie(quantidades, explode=destaque, labels=categorias, autopct='%1.1f%%') # Cria o gr√°fico de pizza 

plt.show() # Exibe o gr√°fico    

```

## Gr√°fico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

fig, ax = plt.subplots()  # Cria a figura e os eixos    
ax.bar(anos, vendas_online, label='online') # Cria o primeiro conjunto de barras
ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja') # Adiciona o segundo conjunto de barras
ax.set_ylabel('Vendas') # Adiciona o r√≥tulo do eixo y   
ax.set_xlabel('ano')  # Adiciona o r√≥tulo do eixo x
ax.set_title('Vendas online vs Lojas') # Adiciona o t√≠tulo do gr√°fico 
ax.set_xticks(anos) # Adiciona os anos no eixo x    
ax.legend() # Adiciona a legenda    
plt.show()
```
