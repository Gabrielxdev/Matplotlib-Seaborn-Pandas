## Personalizando um estilo pr√©-pronto

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

plt.style.use('fivethirtyeight')

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)

plt.show()
```

Todos os par√¢metros dispon√≠veis: https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.rcParams

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

plt.style.use('fivethirtyeight')

params = {
    'axes.titlecolor': 'dimgray', 
    'figure.figsize': (12, 6),
}

plt.rcParams.update(params) 

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos',color='dimgray')
plt.grid(False)

plt.show()
```

## Segundo exemplo

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12, 6))     

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]


ax.plot(meses, vendas, marker='o', color='purple', linestyle='-', linewidth=2, markersize=8)
ax.set_xlabel('Meses')  
ax.set_ylabel('Vendas (em milhares)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')
ax.grid(False)

plt.show()
```
